**Part -1 : Lists, Tuples, Dictionaries, Sets**

```python
# Task 1: Lists
movies = ["Inception", "The Dark Knight", "Interstellar", "Pulp Fiction", "The Matrix"]
print("First:", movies[0], "| Last:", movies[-1])
movies.append("Avatar")
movies.pop(1) # Removes second movie
movies.sort()
print("Sorted List:", movies)
print("Length:", len(movies), "| Exists:", "Inception" in movies)
```

```python
# Task 2: Tuples
cities = ("Chennai", "Mumbai", "Delhi", "Bangalore", "Hyderabad")
print("2nd:", cities[1], "| 4th:", cities[3])
# cities[0] = "Tokyo"  # Error: 'tuple' object does not support item assignment.
# Reason: Tuples are immutable; once created, their elements cannot be changed.
c1, c2, c3, c4, c5 = cities
temp_list = list(cities)
temp_list.append("Kolkata")
cities = tuple(temp_list)
```

```python
# Task 4: Set
weekday_subs = {"Python", "Math", "Stats"}
weekend_subs = {"Stats", "SQL", "Tableau"}
print("Union:", weekday_subs | weekend_subs)
print("Intersection:", weekday_subs & weekend_subs)
print("Difference:", weekday_subs - weekend_subs)
weekday_subs.add("ML")
weekend_subs.discard("SQL")
```

**Part -2 : Functions, Logic and Loops**

```python
# Task 5 & 6: Functions & Default Parameters
def greet(name):
    print(f"Hello, {name}! Welcome to Day 2.")

def add_numbers(a, b):
    return a + b

def find_max(a, b, c):
    return max(a, b, c)

def is_even(n):
    return n % 2 == 0

def introduce(name, language="Python"):
    print(f"{name} is learning {language}.")

def calculate_discount(price, discount=10):
    return price * (1 - discount/100)
```

```python
# Task 7: if / else / elif
num = int(input("Enter a number: "))
if num > 0: print("Positive")
elif num < 0: print("Negative")
else: print("Zero")
```

```python
# Task 8: For Loops
# Sum of 1 to 100
print("Sum 1-100:", sum(range(1, 101)))

# Fruits with index
fruits = ["Apple", "Banana", "Cherry"]
for i, fruit in enumerate(fruits):
    print(f"Index {i}: {fruit}")

# Count vowels
text = "Nunnari Academy"
vowels = sum(1 for char in text.lower() if char in "aeiou")
print(f"Vowel count: {vowels}")
```
